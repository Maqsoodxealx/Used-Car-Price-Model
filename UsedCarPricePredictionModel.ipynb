```python
# Load Dataset from my Google Drive
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/ML Models/Prices_pakwheels_used_cars.csv" )
df.columns = df.columns.str.strip()
df = df.drop(columns=['ad_ref','body','assembly','ad_city'])
df.head(1)
```

```python
def car_price(color, engine_cc, fuel_type, make,
       mileage, model, registered, transmission, year):

    input_df = pd.DataFrame([['color', 'engine_cc', 'fuel_type', 'make','mileage', 'model', 'registered', 'transmission', 'year']])


    le_color = LabelEncoder()
    input_df['color'] = le_color.fit_transform(input_df['color'])
    le_fuel_type = LabelEncoder()
    input_df['fuel_type'] = le_fuel_type.fit_transform(input_df['fuel_type'])
    le_make = LabelEncoder()
    input_df['make'] = le_make.fit_transform(input_df['make'])
    le_model = LabelEncoder()
    input_df['model'] = le_model.fit_transform(input_df['model'])
    le_registered = LabelEncoder()
    input_df['registered'] = le_registered.fit_transform(input_df['registered'])
    le_transmission = LabelEncoder()
    input_df['transmission'] = le_transmission.fit_transform(input_df['transmission'])

    prediction = model.predict(input_df)


iface = gr.Interface(
    fn = car_price,

    inputs=[

        gr.Text(label = 'Enter Car Color'),
        gr.Number(label = 'Enter Engine Capacity in  CC'),
        gr.Dropdown(['Hybrid' , 'Petrol' , 'Diesel'], label = 'Select Fuel Type of Car'),
        gr.Text(label = "Enter Car's Company Name"),
        gr.Number(label = "Enter Car's Mileage"),
        gr.Text(label = "Enter Car's Model"),
        gr.Text(label = "Enter City name or Registered or Un-Registered or Other"),
        gr.Dropdown(['Manual' , 'Automatic'], label = 'Transmission'),
        gr.Number(label = 'Year')
        ],

    outputs = gr.Textbox(label = 'price'),
    title = "Car Price Prediction App",
    description = "Enter Car Details to Predict Price"
)

iface.launch(share=True, debug=True)
```
